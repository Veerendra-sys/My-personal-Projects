```python
# Fill missing values with median or mean based on column types
df['5G Network Coverage (%)'] = df['5G Network Coverage (%)'].fillna(df['5G Network Coverage (%)'].mean())
df['Number of Startups'] = df['Number of Startups'].fillna(df['Number of Startups'].median())
```

# Correlation Matrix

```python
# Select only numerical columns for correlation calculation
numerical_cols = df.select_dtypes(include=['number']).columns

# Calculate correlation matrix
correlation_matrix = df[numerical_cols].corr()

# Plotting the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Numerical Indicators')
plt.show()
```


This correlation matrix visualizes the relationships between various numerical indicators related to technology and business growth.

The diagonal values (all 1.0) indicate that each variable is perfectly correlated with itself.

Most correlations are weak, suggesting that no single factor strongly determines another.

There is a slight positive correlation between Market Share and Startups (0.081), meaning increased market share may encourage new businesses.

R&D Investment and Patents Filed (0.007) show a weak link, indicating that patents depend on more than just R&D spending.

Tech Exports and Number of Tech Companies (0.078) suggest that more tech firms slightly contribute to export growth.

Internet Penetration and Innovation Ranking (-0.017) have a weak negative correlation, implying that better internet access alone doesn't boost innovation rankings

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Country', y='R&D Investment (in USD)', palette=['darkgreen', 'red'])
plt.title('R&D Investment Comparison Between China and Japan')
plt.xlabel('Country')
plt.ylabel('R&D Investment (USD Billions)')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='Number of Patents Filed (Annual)', hue='Country', style='Country', markers=True,palette=['darkgreen', 'red'])
plt.title('Number of Patents Filed Over Time (China vs Japan)')
plt.xlabel('Year')
plt.ylabel('Number of Patents Filed')
plt.grid(True)
plt.show()
```

The graph compares the number of patents filed by China and Japan over time, from 2000 to around 2023. China (blue line) shows more fluctuations but an overall increasing trend, whereas Japan (orange dashed line) has relatively stable but slightly declining trends in certain periods. The shaded regions represent confidence intervals or variability in patent filings for both countries. There are instances where both countries experience sharp drops and recoveries, indicating possible external influences such as economic shifts or policy changes.

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='Global Innovation Ranking', hue='Country', marker="o", palette=['darkgreen', 'red'])
plt.title('Global Innovation Ranking Over Time')
plt.xlabel('Year')
plt.ylabel('Innovation Ranking (Lower is Better)')
plt.show()
```

```python
sector_share = df.groupby(['Tech Sector', 'Country'])['Market Share (%)'].mean().unstack()
sector_share.plot(kind='bar', stacked=True, figsize=(10, 6),color=['darkgreen', 'red'])
plt.title('Market Share Distribution by Tech Sector (China vs Japan)')
plt.ylabel('Average Market Share (%)')
plt.legend(title='Country')
plt.xticks(rotation=45)
plt.show()
```

This stacked bar chart compares the average market share (%) of China and Japan across different tech sectors. China (light blue) and Japan (brown) contribute to the total market share in AI, Biotechnology, Cloud Computing, Robotics, Semiconductors, Software, and Telecommunications. The bars show that both countries have comparable shares in each sector, but the proportion varies.

For example, China has a slightly higher share in Telecommunications and Robotics, while Japan has a stronger presence in Semiconductors and Software. This visualization helps in understanding the relative dominance of each country in different technological fields

```python
china_market_share = df[df['Country'] == 'China'].groupby('Tech Sector')['Market Share (%)'].mean()
colors = sns.color_palette('coolwarm', len(china_market_share))
colors[0] = 'darkgreen'  # Replace first color with dark green
colors[1] = 'red'  # Replace second color with red
plt.figure(figsize=(8, 8))
plt.pie(china_market_share, labels=china_market_share.index, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('China: Market Share by Tech Sector')
plt.show()
```

The chart shows that Telecommunications (16.1%) holds the largest market share in China's tech industry, followed by Robotics (15.3%) and Biotechnology (14.6%). AI and Cloud Computing have slightly lower shares at around 13.5% and 13.4%, respectively. The distribution indicates a strong focus on industrial automation, biotech, and communication technology in China.


```python
japan_market_share = df[df['Country'] == 'Japan'].groupby('Tech Sector')['Market Share (%)'].mean()

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(japan_market_share, labels=japan_market_share.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set3', len(japan_market_share)))
plt.title('Japan: Market Share by Tech Sector')
plt.show()
```

In Japan, Software (15.8%) has the highest market share, followed closely by Semiconductors and Telecommunications (14.3%). Cloud Computing, Robotics, and Biotechnology hold nearly equal shares, reflecting a more balanced investment across different sectors. Japan's focus on Software and Semiconductors suggests an emphasis on IT infrastructure and electronic component development.

Summary Report: China vs. Japan

Overview

This comparative study explores technological progress in China and Japan, analyzing key indicators such as R&D funding, patent activity, sector-wise market share, and digital infrastructure. The dataset covers multiple years, offering a comprehensive perspective on both nations' innovation ecosystems.

Key Insights

1. R&D Expenditure

Chinas R&D investment surpasses Japans, reflecting its aggressive push towards technological dominance.

Japan adopts a strategic, sector-focused R&D approach, excelling in areas such as robotics and material sciences.

2. Patent Activity

Japan leads in the total number of patents filed annually, underscoring its strong emphasis on protecting intellectual property.

China is witnessing rapid growth in patent filings, signaling increasing innovation and research outputs.

3. Global Innovation Index

Japan consistently secures a top position in global innovation rankings due to its well-established research ecosystem.

China is rising steadily in innovation indices, supported by strong investments and infrastructure improvements.

4. Tech Sector Market Share

China dominates in telecommunications and consumer electronics, leveraging its large-scale production capabilities.

Japan leads in robotics and high-precision machinery, carving a niche in specialized high-tech industries.

5. Internet Penetration

Japan exhibits higher internet penetration, demonstrating its superior digital infrastructure and accessibility.

China is rapidly expanding its internet connectivity, driven by government initiatives and urban development.

6. 5G Network Expansion

China is at the forefront of 5G deployment, outpacing Japan in coverage and adoption.

Japan, while progressing, has a gradual 5G rollout compared to Chinas large-scale implementation.

7. Venture Capital Investment

China attracts significant venture capital, fueling a dynamic startup ecosystem and technological innovation.

Japan follows a conservative venture capital model, with a steady but controlled investment approach.

8. Startup Ecosystem

China boasts a thriving startup culture, backed by government support and private investments.

Japan has a smaller but highly specialized startup environment, focused on niche innovations.

9. University Research Collaborations

China is strengthening university-industry partnerships, enhancing research and technological advancements.

Japan maintains a well-established research network, prioritizing long-term collaborations with academic institutions.

10. Leading Tech Exports

China exports a wide range of tech products, including consumer electronics, telecom equipment, and IT devices.

Japans exports are high-value tech components, such as semiconductors, robotics, and precision instruments.

11. Tech Workforce Strength

China has an extensive tech workforce, aligning with its large-scale industrial growth.

Japan, while having a smaller workforce, is recognized for high expertise and specialization.

Visual Insights

Correlation Analysis: A strong correlation exists between R&D spending and patent filings, emphasizing the impact of financial backing on innovation.

Trend Analysis: Over time, Chinas R&D and patent growth show rapid acceleration, whereas Japan maintains a steady trajectory.

Market Distribution: Graphs and pie charts highlight the sector-wise dominance of each country, illustrating their respective strengths.

Conclusion

The study highlights how China is evolving as a global tech powerhouse, leveraging massive investments and large-scale innovation, while Japan remains a leader in high-quality, specialized technologies. Both nations play vital roles in the international tech landscapeChina driving expansion and accessibility, and Japan ensuring precision and sustainability in technological advancements.
