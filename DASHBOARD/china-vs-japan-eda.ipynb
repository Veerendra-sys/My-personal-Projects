```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import numpy as np
import warnings

# To display visuals within the notebook
%matplotlib inline
warnings.filterwarnings('ignore')
```

<a id="2"></a>
# <div style="text-align:center; border-radius:15px 50px; padding:15px; color:white; margin:0; font-size:100%; font-family:Arial; background-color:green; overflow:hidden"><b> ‚ú® Load the Dataset ‚ú®</b></div>

```python
# Load the dataset from the provided file path
file_path = '/kaggle/input/china-vs-japan/Big_Japan_vs_China_Technology.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
df.head(10)
```

<a id="2"></a>
# <div style="text-align:center; border-radius:15px 50px; padding:15px; color:white; margin:0; font-size:100%; font-family:Arial; background-color:green; overflow:hidden"><b> ‚ú® Basic Dataset Overview ‚ú®</b></div>

```python
# Get an overview of the dataset
df.info()

# Summary statistics for numerical columns
df.describe()
```

<a id="2"></a>
# <div style="text-align:center; border-radius:15px 50px; padding:15px; color:white; margin:0; font-size:100%; font-family:Arial; background-color:green; overflow:hidden"><b> ‚ú® Checking for Missing Values ‚ú®</b></div>

```python
# Check for any missing values
missing_values = df.isnull().sum()
missing_values[missing_values > 0]  # Display columns with missing values
```

<a id="2"></a>
# <div style="text-align:center; border-radius:15px 50px; padding:15px; color:white; margin:0; font-size:100%; font-family:Arial; background-color:green; overflow:hidden"><b> ‚ú® Handling Missing Values ‚ú®</b></div>

```python
# Fill missing values with median or mean based on column types
df['5G Network Coverage (%)'] = df['5G Network Coverage (%)'].fillna(df['5G Network Coverage (%)'].mean())
df['Number of Startups'] = df['Number of Startups'].fillna(df['Number of Startups'].median())

# Verify missing values have been handled
df.isnull().sum().sum()  # Should return 0 if no missing values remain
```

<a id="2"></a>
# <div style="text-align:center; border-radius:15px 50px; padding:15px; color:white; margin:0; font-size:100%; font-family:Arial; background-color:green; overflow:hidden"><b> ‚ú® Correlation Matrix ‚ú®</b></div>

```python
# Select only numerical columns for correlation calculation
numerical_cols = df.select_dtypes(include=['number']).columns

# Calculate correlation matrix
correlation_matrix = df[numerical_cols].corr()

# Plotting the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Numerical Indicators')
plt.show()
```

<a id="2"></a>
# <div style="text-align:center; border-radius:15px 50px; padding:15px; color:white; margin:0; font-size:100%; font-family:Arial; background-color:green; overflow:hidden"><b> ‚ú® Visualizing R&D Investment by Country ‚ú®</b></div>

```python
# R&D Investment Comparison
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Country', y='R&D Investment (in USD)', palette='Set2')
plt.title('R&D Investment Comparison Between China and Japan')
plt.xlabel('Country')
plt.ylabel('R&D Investment (USD Billions)')
plt.show()
```

<a id="2"></a>
# <div style="text-align:center; border-radius:15px 50px; padding:15px; color:white; margin:0; font-size:100%; font-family:Arial; background-color:green; overflow:hidden"><b> ‚ú® Number of Patents Filed Over Time ‚ú®</b></div>

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='Number of Patents Filed (Annual)', hue='Country', style='Country', markers=True)
plt.title('Number of Patents Filed Over Time (China vs Japan)')
plt.xlabel('Year')
plt.ylabel('Number of Patents Filed')
plt.grid(True)
plt.show()
```

<a id="2"></a>
# <div style="text-align:center; border-radius:15px 50px; padding:15px; color:white; margin:0; font-size:100%; font-family:Arial; background-color:green; overflow:hidden"><b> ‚ú® Global Innovation Ranking Over Time ‚ú®</b></div>

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='Global Innovation Ranking', hue='Country', marker="o", palette='Dark2')
plt.title('Global Innovation Ranking Over Time')
plt.xlabel('Year')
plt.ylabel('Innovation Ranking (Lower is Better)')
plt.show()
```

<a id="2"></a>
# <div style="text-align:center; border-radius:15px 50px; padding:15px; color:white; margin:0; font-size:100%; font-family:Arial; background-color:green; overflow:hidden"><b> ‚ú® Market Share Distribution by Tech Sector ‚ú®</b></div>

```python
# Group by Tech Sector and Country
sector_share = df.groupby(['Tech Sector', 'Country'])['Market Share (%)'].mean().unstack()

# Plot stacked bar chart
sector_share.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='Paired')
plt.title('Market Share Distribution by Tech Sector (China vs Japan)')
plt.ylabel('Average Market Share (%)')
plt.legend(title='Country')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='Internet Penetration (%)', hue='Country', style='Country', markers=True, dashes=False)
plt.title('Internet Penetration Over Time (China vs Japan)')
plt.xlabel('Year')
plt.ylabel('Internet Penetration (%)')
plt.grid(True)
plt.show()
```

# Summary Report: China vs. Japan

## Overview
This analysis provides an in-depth comparison of technological advancements between China and Japan based on several metrics including R&D investment, patent filings, market share by tech sector, and more. The dataset spans multiple years and includes diverse aspects of technology and innovation.

## Key Findings

### 1. R&D Investment
- **China** significantly outspends Japan in R&D investments. This substantial investment underpins China‚Äôs rapid advancements in technology and innovation.
- **Japan** has a more focused R&D strategy, often leading in specific tech sectors like robotics and advanced materials.

### 2. Number of Patents Filed
- **Japan** has a higher number of patents filed annually compared to China. This indicates Japan's strong emphasis on technological innovation and intellectual property.
- **China** is making substantial progress in patent filings, reflecting its growing focus on research and development.

### 3. Global Innovation Ranking
- **Japan** consistently ranks higher in global innovation indices. This reflects its strong historical performance in technology and innovation.
- **China** is climbing the ranks rapidly, thanks to its significant investments in technology and infrastructure.

### 4. Market Share by Tech Sector
- **China** holds a dominant market share in sectors like telecommunications and electronics, benefiting from its large-scale manufacturing capabilities.
- **Japan** leads in high-tech sectors such as robotics and precision machinery, where it has established a niche due to advanced technology and quality.

### 5. Internet Penetration
- **Japan** has a higher rate of internet penetration compared to China. This indicates better infrastructure and accessibility in Japan.
- **China** is making significant strides, with increasing internet access and digital infrastructure improvements.

### 6. 5G Network Coverage
- **China** leads in 5G network coverage, showcasing its advancements in telecommunications and rapid deployment of next-generation networks.
- **Japan** is also progressing in 5G but has a slower rollout compared to China.

### 7. Venture Capital Funding
- **China** receives higher venture capital funding, reflecting its vibrant startup ecosystem and investor confidence in its tech sector.
- **Japan** has a more conservative venture capital environment but is growing steadily.

### 8. Number of Startups
- **China** has a larger number of startups, driven by its dynamic tech ecosystem and supportive policies for entrepreneurship.
- **Japan** has fewer startups but they are often highly specialized and innovative.

### 9. University Research Collaborations
- **China** shows increasing collaborations with universities, highlighting its commitment to research and development through academic partnerships.
- **Japan** maintains strong collaborations but with a focus on established research institutions and long-term partnerships.

### 10. Top Tech Products Exported
- **China** exports a broad range of tech products including consumer electronics, telecommunications equipment, and IT hardware.
- **Japan** is known for exporting high-value tech products such as semiconductors, robotics, and precision instruments.

### 11. Number of Tech Workers
- **China** has a significantly larger number of tech workers, supporting its vast tech industry and rapid expansion.
- **Japan** has a highly skilled tech workforce, but its numbers are smaller due to a more mature and focused industry.

## Visual Insights
- **Correlation Analysis**: The correlation matrix reveals strong relationships between R&D investment and number of patents filed, underscoring the link between financial investment and innovation output.
- **Time Trends**: Trends in R&D investment and patent filings over time show China's rapid growth and Japan's steady performance.
- **Sectoral Distribution**: Pie charts and bar graphs illustrate the dominance of specific tech sectors in each country, providing a clear picture of market share distribution.

## Conclusion
The analysis reveals a dynamic technological landscape where China is rapidly advancing with significant investments and broad market impact, while Japan continues to excel in innovation and specialized tech sectors. Both countries contribute uniquely to the global technology ecosystem, with China driving large-scale growth and Japan focusing on high-quality and specialized technological advancements.
